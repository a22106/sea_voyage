```python
import searoute as sr
from utils.route_utils import make_searoute_edges, make_searoute_nodes
from utils.map_utils import map_folium

start_point = (129.170, 35.075)
end_point = (-4.158630, 44.644694)
lat1 = 33.618546
lon1 = 126.493268
lat2 = 33.668537
lon2 = 127.153811

# 새로운 노드 추가
new_node1 = (lon1, lat1)
new_node2 = (lon2, lat2)
nodes = []
nodes.append(new_node1)
nodes.append(new_node2)
updated_nodes = make_searoute_nodes(nodes)

from searoute.data.marnet_dict import edge_list as marnet_e, node_list as marnet_n
current_marnet = sr.from_nodes_edges_set(sr.Marnet(), marnet_n, marnet_e)
updated_edges = make_searoute_edges(updated_nodes, current_marnet)
current_marnet.add_nodes_from(updated_nodes)
current_marnet.add_edges_from(updated_edges)

# 경로 생성
route = sr.searoute(start_point, end_point, M=current_marnet)
map_folium(route["geometry"])
```
